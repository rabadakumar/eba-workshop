```python
#dict - Python Dictionary Operations - Key : value

# Create Dictionary
student = {
    "name": "shalini",
    "age": 20,
    "course": "Python"
}

# Access Value by key
print(student["age"])

# Add New Item - both key and value
student["grade"] = "A"
student

```

```python
# Update Value by key
student["age"] = 21

# Remove Item by key
student.pop("course")
student

```

```python
# Get Keys
print(student.keys())

# Get Values
print(student.values())

# Get Items
print(student.items())

# Check Key
print("name1" in student)


```

```python
# Dictionary Length
print(len(student))

```

```python
# Clear Dictionary
student.clear()

print(student)
```

```python
#use define function - created by user
#with parameter

def addition(a,b):  #a , b are 2 parameter
  result=a+b
  print(result)

def subtraction(a,b):
  result=a-b
  print(result)

#call the function
addition(5,6).  #passing the value to parameters
```

```python
#without parameter
def addition(): #no parameter inside
  result=5+8
  print(result)
addition()
```

```python
#lambda function - 1 line expression
addition= lambda a,b:a+b
addition(3,6)
```

```python
square=lambda x:x*x
square(5)
```
